```python
# 1. Giải nén bộ dữ liệu bạn vừa tải lên
!unzip "car scratch.v1i.yolov8.zip" -d "car_dataset"

# 2. Cài đặt bộ công cụ AI YOLOv8
!pip install ultralytics

# 3. Khởi động AI
from ultralytics import YOLO
model = YOLO('yolov8n.pt') # Tải "bộ não" AI cơ bản, nhẹ nhất

# 4. Cho AI bắt đầu đi học (Train)
# epochs=15 nghĩa là nó sẽ đọc đi đọc lại bộ dữ liệu 15 lần để ghi nhớ vết xước
results = model.train(data='car_dataset/data.yaml', epochs=15, imgsz=640)
```

```python
from ultralytics import YOLO
from IPython.display import Image, display

# 1. Lấy "bộ não" giỏi nhất mà AI vừa học được lúc nãy ra dùng
model = YOLO('runs/detect/train/weights/best.pt')

# 2. Bắt nó nhìn vào ảnh test.jpg và khoanh vùng vết xước
# save=True nghĩa là yêu cầu nó vẽ khung màu và lưu lại thành ảnh mới
results = model.predict(source='test.jpg', save=True)

# 3. Hiển thị bức ảnh kết quả ngay trên màn hình này cho bạn xem
duong_dan_anh_ket_qua = f"{results[0].save_dir}/test.jpg"
display(Image(filename=duong_dan_anh_ket_qua))
```
